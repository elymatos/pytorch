```python
import numpy as np
import matplotlib.pyplot as plt

# === Cortical-Level Simulation of HTPC in Recognition Mode ===

# Sentence to recognize (tokenized)
sequence = ['the', 'dog', 'ran']
n_tokens = len(sequence)
n_levels = 2  # Layer hierarchy: Level 0 = local tokens, Level 1 = phrase
n_trials = 10

# Storage for predictions and errors per level
level_predictions = [np.zeros(n_tokens) for _ in range(n_levels)]
prediction_errors = [[] for _ in range(n_levels)]

# === Simulated learned prediction for temporal (token) level ===
# -------------------------------
#  L5 at time t-1 predicts what L4 at time t will receive
# These are pre-trained predictions (no learning here)
token_prediction_matrix = {
    'the': 'dog',
    'dog': 'ran',
    'ran': None
}

# === Simulated top-down prediction from high-order column ===
# -------------------------------
#  L5 in higher column predicts entire sentence
# Delivered via L1 to L2/3 across time
phrase_prediction_vector = np.array([1, 1, 1])  # All tokens are expected

# === Recognition trials loop ===
for trial in range(n_trials):
    for i, token in enumerate(sequence):

        # === Layer 4: Receives bottom-up input from thalamus ===
        actual = 1.0  # Token arrives at L4

        # === Layer 5  L1: Generate prediction from previous step ===
        predicted_token = None
        if i > 0:
            prev_token = sequence[i - 1]
            predicted_token = token_prediction_matrix.get(prev_token, None)

        # === Layer 2/3: Compare prediction and actual input ===
        # Prediction arrives from L5 (apical), input from L4 (basal)
        low_pred = 1.0 if predicted_token == token else 0.0
        error = abs(actual - low_pred)
        level_predictions[0][i] = low_pred
        prediction_errors[0].append(error)

        # === Higher-Order Column L5: Predict global structure ===
        # This is contextual/semantic expectation (e.g., PFC)
        high_pred = phrase_prediction_vector[i]
        error_high = abs(actual - high_pred)
        level_predictions[1][i] = high_pred
        prediction_errors[1].append(error_high)

# === Plotting prediction errors for each level ===
fig, ax = plt.subplots(n_levels, 1, figsize=(10, 6), sharex=True)

labels = ['Low-Level (L2/3: Token)', 'High-Level (L5: Phrase)']
for lvl in range(n_levels):
    ax[lvl].plot(prediction_errors[lvl], label=labels[lvl])
    ax[lvl].set_ylabel("Prediction Error")
    ax[lvl].legend()
    ax[lvl].grid(True)

ax[-1].set_xlabel("Recognition Step (Tokens over Trials)")
plt.suptitle("HTPC Recognition Mode: Cortical Flow and Prediction Errors")
plt.tight_layout()
plt.show()
```
